```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charger les données depuis le fichier CSV (assurez-vous d'ajuster le chemin)
data = pd.read_csv(r'C:\Users\anask\Desktop\AnalyzeDatascrapping\data.csv')
```

```python
# Afficher les premières lignes du dataframe pour comprendre la structure des données
print(data.head())
```

```python
# 1. Nombre d'applications par catégorie
category_counts = data['Catégorie'].value_counts()
category_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Nombre d\'applications par catégorie')
plt.xlabel('Catégorie')
plt.ylabel('Nombre d\'applications')
plt.show()
```

```python
# 2. Prix moyen par catégorie
avg_price_by_category = data.groupby('Catégorie')['Prix'].mean()
avg_price_by_category.plot(kind='bar', figsize=(10, 6), color='salmon')
plt.title('Prix moyen par catégorie')
plt.xlabel('Catégorie')
plt.ylabel('Prix moyen')
plt.show()
```

```python
# 3. Répartition des développeurs
developer_counts = data['Developpeur'].value_counts().head(10)
developer_counts.plot(kind='bar', figsize=(12, 6), color='lightgreen')
plt.title('Top 10 des développeurs par nombre d\'applications')
plt.xlabel('Developpeur')
plt.ylabel('Nombre d\'applications')
plt.show()
```

```python
# 4. Date de sortie des applications
data['Date De Sortie'] = pd.to_datetime(data['Date De Sortie'])
monthly_release_counts = data.resample('M', on='Date De Sortie').size()
monthly_release_counts.plot(figsize=(12, 6), color='orange')
plt.title('Nombre d\'applications sorties chaque mois')
plt.xlabel('Mois de sortie')
plt.ylabel('Nombre d\'applications')
plt.show()
```
